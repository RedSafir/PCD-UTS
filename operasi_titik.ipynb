# Operasi Titik

```python
import cv2 
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

<h3> Load Gambar </h3>

```python
img = cv2.imread("dumy-img-1.jpg")
cv2_imshow(img)
```

<h3> terdapat beberapa operasi yang dimiliki operasi titik </h3>
<p> yaitu : </p>
<ol>
  <li>GreyScale</li>
  <li>Brightnes</li>
  <li>Contras</li>
  <li>Contras Strecth</li>
  <li>Negative</li>
  <li>Biner</li>
</ol>

#1. GreyScale
<p> f(x)' = 0.299 R + 0.587 G + 0.144 B </p> 

```python
img = cv2.imread("dumy-img-1.jpg")

H, W = img.shape[:2]
grey = np.zeros((H, W))

for i in range(H):
  for j in range(W):
    grey[i,j] = np.clip(0.299 * img[i, j, 0] +
                        0.587 * img[i, j, 1] +
                        0.144 * img[i, j, 2], 0, 255)

cv2_imshow(grey)
```

#2. Brightness
f(x)' = f(x) + b 

```python
img = cv2.imread("dumy-img-1.jpg", 0)

H, W = img.shape[:2]
BRIGHTNESS = 50

for i in range(H):
  for j in range(W):
    a = img.item(i, j)
    img[i, j] = np.clip(a + BRIGHTNESS, 0, 255)

cv2_imshow(img)
```

#3. Contras
<p> f(x)' = f(x) * C

```python
img = cv2.imread("dumy-img-1.jpg", 0)

H, W = img.shape[:2]
CONTRAS = 0.5

for i in range(H):
  for j in range(W):
    a = img.item(i, j)
    img[i, j] = np.clip(a * CONTRAS, 0, 255)

cv2_imshow(img)
```

#4. Contras Streching
<ol>
  <li>Cari nilai maks dan min pada pixel</li>
  <li>f(x)' = (f(x) - min) / (max - min) * 255

```python
img = cv2.imread("dumy-img-1.jpg", 0)

H, W = img.shape[:2]
minG = np.min(img)
maxG = np.max(img)

for i in range(H):
  for j in range(W):
    a = img.item(i, j)
    img[i, j] = np.clip(int((a - minG) / (maxG - minG) * 255), 0, 255)

cv2_imshow(img)
```

#5. Binner
<p> f(x)' = 255 - f(x)</p>

```python
img = cv2.imread("dumy-img-1.jpg", 0)

H, W = img.shape[:2]

for i in range(H):
  for j in range(W):
    a = img.item(i, j)
    img[i, j] = np.clip(255 - img[i, j], 0, 255)

cv2_imshow(img)
```
